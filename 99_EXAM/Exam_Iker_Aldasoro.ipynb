# **2. REGRESSION**

```python
import numpy as np
import math
```

```python
# Define the file path
file_path = '/content/drive/My Drive/01_1ST_COURSE/01_DATOS/11302024_EXAMEN/100538243.npz'

# Read the CSV file into a pandas DataFrame
data =  np.load(file_path)
variable1 = data["xRtrain"]
variable2 = data["xRval"]
variable3 = data["sRtrain"]
variable4 = data["sRval"]

# Display the first few rows of the DataFrame
#print(variable1)
#print(variable2)
#print(variable3)
#print(variable4)
```

EXERCICE 1

```python
col1 = np.ones( (300,) )
col2 = np.ones( (100,) )
print(col1.shape)
print(variable1.shape)

L2train = np.linalg.norm(variable1, axis=1)
print(L2train.shape)
L2val = np.linalg.norm(variable2, axis=1)
print(L2val.shape)


test = np.vstack((L2train, col1))
test = np.vstack((test, col1))
test = np.vstack((test, col1))
L2train = np.vstack((test, col1)).T

test2 = np.vstack((L2val, col2))
test2 = np.vstack((test2, col2))
test2 = np.vstack((test2, col2))
L2val = np.vstack((test2, col2)).T
print(L2val.shape)
```

```python
xUtrain = variable1/L2train
#print(xUtrain)
print(xUtrain.shape)
xUval = variable2/L2val
#print(xUval)
print(xUval.shape)
```

EXERCICE 2

```python
sigma_p = np.sqrt(2)
sigma_eps = np.sqrt(0.2)

def posterior_stats(Z, s, sigma_eps, sigma_p):
  dim_w = Z.shape[1]
  iCov_w = (Z.T @ Z)/(sigma_eps**2) + np.eye(dim_w, dim_w)/(sigma_p**2)
  Cov_w = np.linalg.inv(iCov_w)
  mean_w = (Cov_w @ Z.T @ s)/(sigma_eps**2)
  return mean_w, Cov_w, iCov_w

w108, V108, _ = posterior_stats(xUtrain, variable3, sigma_eps,sigma_p)
print(w108)
print(V108)
```

EXERCICE 3

```python
m_s = xUtrain.dot(w108)
MSE = np.mean((m_s - variable3)**2)
RECPtrain = math.sqrt(MSE)

print("Root Mean Square Error:\n")
print(RECPtrain)
```
